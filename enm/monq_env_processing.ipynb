## Clip future WorldClim data using corresponding buffer

```python
# Clip future WorldClim data
year_list = ["2011-2040","2041-2070","2071-2100"]
years_list = ["30y","60y","90y"]
scenario_list = ["ssp126"]
bioclim_list = ["bio2","bio5","bio7","bio8","bio9","bio14","bio15","bio18","scd","swe"]
topovar_list = ["aspect","slope"]
distance_list = [150,300,450]

futureclim_folder = f"{project_folder}/bioclim/future/GLOBAL/climatologies"

for i in range(len(year_list)):
        year = year_list[i]
        years = years_list[i]
        distance = distance_list[i]
        print("clipping " + year + " bioclim data by " + str(distance) + " km")
        clipFeature = f"{arcpy.env.workspace}/monq_{distance}km_{years}_onLand.shp"
        outFolder = f"{project_folder}/bioclim/future/processed/{year}"

        for scen in scenario_list:
            for var in bioclim_list:
                inRaster = f"{futureclim_folder}/{year}/GFDL-ESM4/{scen}/bio/CHELSA_{var}_{year}_gfdl-esm4_{scen}_V.2.1.tif"
                outRaster = f"{outFolder}/{var}.tif"
                clip_raster = arcpy.ia.Clip(inRaster, aoi = clipFeature)
                clip_raster.save(outRaster)
            for var in topovar_list:
                inRaster = f"{project_folder}/bioclim/monq_dem/monq_{var}_res_clip.tif"
                outRaster = f"{outFolder}/{var}.tif"
                clip_raster = arcpy.ia.Clip(inRaster, aoi = clipFeature)
                clip_raster.save(outRaster)
```
